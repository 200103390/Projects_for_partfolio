```python
import folium

map = folium.Map(location=[43.187805,76.886012], zoom_start = 11)

map
```

```python
import pandas as pd
import math
data = pd.read_csv("Data.csv")
lat = data['LAT']
lon = data['LON']
elevation = data['Name']
num1 = []
num2 = []
num3 = []
num4 = []
lnum1 = []
lnum2 = []
lnum3 = []
lnum4 = []

for lat, lon, elevation in zip(lat, lon, elevation):
    if lat>=43.25887888074892 and lat<=43.26827881162943 and lon>=76.952884767155 and lon<=76.976536530374:
        num1.append(lat)
        lnum1.append(lon)
        folium.Marker(location=[lat, lon], popup=str(elevation)+" m", icon=folium.Icon(color = 'red')).add_to(map)
    elif lat>=43.192286830212396  and lon<=76.9127006607541 :
        num2.append(lat)
        lnum2.append(lon)
        folium.Marker(location=[lat, lon], popup=str(elevation)+" m", icon=folium.Icon(color = 'blue')).add_to(map)
    elif lat<=43.161828821070685:
        num3.append(lat)
        lnum3.append(lon)
        folium.Marker(location=[lat, lon], popup=str(elevation)+" m", icon=folium.Icon(color = 'pink')).add_to(map)
    else:
        num4.append(lat)
        lnum4.append(lon)
        folium.Marker(location=[lat, lon], popup=str(elevation)+" m", icon=folium.Icon(color = 'green')).add_to(map)
l1 = 0
l11 = 0
l2 = 0
l22 = 0
l3 = 0
l33 = 0
l4 = 0
l44 = 0
for i in range(0,len(num1)):
    l1 = l1 + num1[i];
    l11 = l11 + lnum1[i];
for i in range(0,len(num2)):
    l2 = l2 + num2[i];
    l22 = l22 + lnum2[i]; 
for i in range(0,len(num3)):
    l3 = l3 + num3[i];
    l33 = l33 + lnum3[i];
for i in range(0,len(num4)):
    l4 = l4 + num4[i];
    l44 = l44 + lnum4[i];  
    
centroid1_lon = l1/len(num1);    
centroid1_lat = l11/len(num1); 
centroid2_lon = l2/len(num2);    
centroid2_lat = l22/len(num2);
centroid3_lon = l3/len(num3);    
centroid3_lat = l33/len(num3); 
centroid4_lon = l4/len(num4);    
centroid4_lat = l44/len(num4); 

folium.Marker(location=[centroid1_lon, centroid1_lat], popup=str(elevation)+" m", icon=folium.Icon(color = 'gray')).add_to(map)
folium.Marker(location=[centroid2_lon, centroid2_lat], popup=str(elevation)+" m", icon=folium.Icon(color = 'gray')).add_to(map)
folium.Marker(location=[centroid3_lon, centroid3_lat], popup=str(elevation)+" m", icon=folium.Icon(color = 'gray')).add_to(map)
folium.Marker(location=[centroid4_lon, centroid4_lat], popup=str(elevation)+" m", icon=folium.Icon(color = 'gray')).add_to(map)
dist = 0 
dist1 = 0
dist2 = 0 
dist3 = 0

for i in range(0,len(num1)):
    dist = dist + math.sqrt(pow(centroid1_lon-num1[i] , 2) + pow(centroid1_lat-lnum1[i] , 2))
for i in range(0,len(num2)):
    dist1 = dist1 + math.sqrt(pow(centroid2_lon-num2[i] , 2) + pow(centroid2_lat-lnum2[i] , 2))    
for i in range(0,len(num3)):
    dist2 = dist2 + math.sqrt(pow(centroid3_lon-num3[i] , 2) + pow(centroid3_lat-lnum3[i] , 2))
for i in range(0,len(num4)):
    dist3 = dist3 + math.sqrt(pow(centroid4_lon-num4[i] , 2) + pow(centroid4_lat-lnum4[i] , 2))    
        
dist = dist/len(num1);
dist1 = dist1/len(num2);
dist2 = dist2/len(num3);
dist3 = dist3/len(num4);

if dist<dist1 and dist<dist2 and dist<dist3:
    print("Best cluster is red")
elif dist>dist1 and dist1<dist2 and dist1<dist3:
    print("Best cluster is blue")
elif dist>dist2 and dist1>dist2 and dist2>dist3:
    print("Best cluster is pink")
else:
    print("Best cluster is green")

    
map
```
