```python
"""
My task is finding best place for tourist which want come to Almaty, and want to find, where he(she) can visit max number of 
places on foot.
"""

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
import folium 

data = pd.read_csv("data_tourists.csv" , sep = ';')

# Generate some synthetic data
```

```python
X = data[["Lat" , "Lon"]].values
PlaceName = data['PlaceName']
```

```python
# Specify the number of clusters (k)
num_clusters = 3

# Create a KMeans instance
kmeans = KMeans(n_clusters=num_clusters)

# Fit the KMeans model to the data
kmeans.fit(X)

# Get the cluster assignments and cluster centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
#Colors for each of cluster
cluster_colors = ['blue', 'green', 'purple']
# Visualize the data and clusters
map_center = X.mean(axis=0)
m = folium.Map(location=map_center, zoom_start=10)

#Calculate the  distances between centroids and points 
distances = np.linalg.norm(X - centroids[labels], axis=1)



#Calculate the avarage distances between centroids and points 
cluster_avg_distances = []
for cluster_idx in range(num_clusters):
    cluster_points = X[labels == cluster_idx]
    avg_distance = np.mean(distances[labels == cluster_idx])
    cluster_avg_distances.append(avg_distance)

    
for i in cluster_avg_distances:
    print(i)    
# Find the cluster with the smallest average distance
smallest_avg_distance_cluster = np.argmin(cluster_avg_distances)
    
```

```python
for i, (lat, lon) in enumerate(X):
    cluster_color = cluster_colors[labels[i]]
    folium.Marker([lat, lon], popup=f'Data Point {i}', icon=folium.Icon(color=cluster_color)).add_to(m)

# Add markers for cluster centroids
for i, (lat, lon) in enumerate(centroids):
    folium.Marker([lat, lon], popup=f'Centroid {i}', icon=folium.Icon(color='red')).add_to(m)


#Mark the best centroid with gray color
smallest_avg_distance_centroid = centroids[smallest_avg_distance_cluster]
folium.Marker(smallest_avg_distance_centroid, popup=f'Smallest Distance Cluster', icon=folium.Icon(color='gray')).add_to(m)
    
# Show the map

m
```
